#### 1. Use el m칠todo de bisecci칩n para encontrar soluciones precisas dentro de $10^{-2}$ para $x^{3}-7x^{2}+14x-6=0$ en cada intervalo

```python
def bisection_method(func, a, b, tol=1e-5):
    """
    M칠todo de bisecci칩n para encontrar una ra칤z de una funci칩n en un intervalo [a, b] con una tolerancia tol.

    Parameters:
    func (function): La funci칩n para la cual se desea encontrar una ra칤z.
    a (float): El extremo izquierdo del intervalo.
    b (float): El extremo derecho del intervalo.
    tol (float): La tolerancia para la aproximaci칩n de la ra칤z.

    Returns:
    float: La aproximaci칩n de la ra칤z.
    """
    if func(a) * func(b) >= 0:
        print("La funci칩n no cambia de signo en el intervalo dado.")
        return None

    c = a
    while (b - a) >= tol:
        c = (a + b) / 2
        if func(c) == 0.0:
            break
        if func(c) * func(a) < 0:
            b = c
        else:
            a = c
    return c
```

```python
def f(x):
    return x**3 - 7*x**2 + 14*x - 6
```

$a. [0, 1]$

```python
a, b = 0, 1
root = bisection_method(f,a, b)
print(f"La ra칤z en el intervalo [{a}, {b}] es: {root}")
```

$b. [1, 3.2]$

```python
a, b = 1, 3.2
root = bisection_method(f,a, b)
print(f"La ra칤z en el intervalo [{a}, {b}] es: {root}")
```

$c. [3.2, 4]$

```python
a, b = 3.2, 4
root = bisection_method(f,a, b)
print(f"La ra칤z en el intervalo [{a}, {b}] es: {root}")
```

#### 1. Un abrevadero de longitud $L$ tiene una secci칩n transversal en forma de semic칤rculo con radio $r$. (Consulte la figura adjunta.). Cuando sellena con agua hasta una distancia $h$ a partir de la parte superior, el volumen $V$ de agua
$$V=L[0.5\pi r^2 - r^2 arcsen(h/r)-h(r^2-h^2)^{1/2}]$$

![Ejercicios Aplicados 1](aplicados1.PNG)

#### Suponga que $L = 10 cm, r = 1 cm$ y $V =12.4 cm$. Encuentre la profundidad del agua en el abrevadero dentro de $0.01 cm$ 

```python
def V(h, L=10, r=1):
    
    if h < -r or h > r:
        return np.nan
    try:
        return L * (0.5 * np.pi * r**2 - r**2 * np.arcsin(h/r) - h * np.sqrt(r**2 - h**2)) - 12.4
    except ValueError:
        return np.nan 

a, b = 0, 2  # Intervalo inicial
root = bisection_method(V, a, b, tol=0.01)
print(f"La profundidad del agua es: {root} cm")
```

#### 2. Un objeto que cae verticalmente a trav칠s del aire est치 sujeto a una resistencia viscosa, as칤 como a la fuerza de gravedad. Suponga que un objeto con masa $m$ cae desde una altura $s_0$ y que la altura del objeto despu칠s de $t$ segundos es $$s(t)=s_0 - \frac{mg}{k}t + \frac{m^2g}{k^2}(1-e^{-kt/m}),$$ donde $g=9.81m/s^2$ y $k$ representa el coeficiente de la resistencia del aire en $Ns/m$. Suponga $s_0=300m$, $m=0.25kg$ y $k=0.1Ns/m$. Encuentre dentro de $0.01\space segundos$, el tiempo que tarda un cuarto de kg en golpear el piso.

```python
g = 9.81
m = 0.25
k = 0.1
s0 = 300

def s(t):
    return s0 - (m/k)*g*t + (m/k)**2 * g * (1 - np.exp(-(k/m)*t))

def height(t):
    return s(t)

a, b = 0, 100  # Intervalo inicial grande
root = bisection_method(height, a, b, tol=0.01)
print(f"El tiempo que tarda en golpear el piso es: {root} segundos")
```
